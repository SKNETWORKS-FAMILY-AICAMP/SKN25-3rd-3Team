```python
from app.db.postgres_db import get_postgres_connection

with get_postgres_connection() as conn:
    with conn.cursor() as cur:
        cur.execute("""
            SELECT DISTINCT pri.raw_text
            FROM parsed_recipe_ingredient AS pri
            WHERE pri.warnings ? 'missing_quantity'
              AND NOT (pri.warnings ? 'tool_detected')
            ORDER BY pri.raw_text
            LIMIT 100
        """)
        rows = cur.fetchall()

[row["raw_text"] for row in rows]

```
